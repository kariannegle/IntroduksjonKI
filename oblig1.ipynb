```python
#Task 1
import pandas as pd

#Read the CSV file inot a DataFrame called GRc_df
Grc_df = pd.read_csv('data/Grocery_dataset.csv')

#Check the number of rows and columns in the DataFrame
print(f"Number of rows and columns: {Grc_df.shape}")

#Show the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(Grc_df.head(10))

#Show the last 10 rows of the DataFrame
print("Last 10 rows of the DataFrame:")
print(Grc_df.tail(10))
```

```python
# Task 2: Check for null values and handle them

# Check for null values and print in a readable format
null_counts = Grc_df.isna().sum()
print("Null values in each column before handling:")
print(null_counts[null_counts > 0])  # Display only columns with null values

# Handle null values by dropping them
Grc_df.dropna(inplace = True)

# Check for null values after handling them
Grc_df.isna().sum()
```
